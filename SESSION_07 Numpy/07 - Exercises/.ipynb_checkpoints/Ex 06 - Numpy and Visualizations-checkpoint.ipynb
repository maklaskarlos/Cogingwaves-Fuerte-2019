## Line plot

2. Create a 1d-array of 30 data points from range 10 to range 40 and use it like the **x** variable. Create the **y** variable by multiply **x * 2**.
 * Plot the data in a line plot

```python
# Creating x and y
x= np.arange(10,40)
y= x*2
```

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x,y)


# Setting the title
plt.title("Karlos Plot")


# Setting the axis names
plt.xlabel("This is the X")
plt.ylabel("This is the Y")

# show
plt.show()
```

3. Use the same data you created at point 2 and plot them by visualizing only **dots** 

```python
# Set the figure size
plt.figure(figsize=(8,4))

# Plotting the data
plt.plot(x,y, "o")

# Setting the title
plt.title("Karlos Garcia")

# Setting the axis names
plt.xlabel("This is the x axis")
plt.ylabel("This is the y axis")
# show
plt.show()
```

### Exponential plot

4. Use the data you have created at point 2, this time the **y** variable must be the exponential **exp()** of the **x**.
 * Plot them in a line plot

```python
# Calculate exponential for all elements in the array.
x= np.arange(10,40)
y= x**2
```

```python
# Make a new plot
plt.figure(figsize=(8,4))

# plot function
plt.plot(x,y)

# setting the title
plt.title("Karlos Garcia")

# setting axis
plt.xlabel("this is the x axis")
plt.ylabel("this is the y axis")
# show
plt.show()
```

### Multiple lines on the same plot

5. Use the data belowe. Plot them on the same plot, use color red for **y1** and green for **y2**

```python
# data
x = np.arange(1,11)
y1 = x ** 2
y2 = x ** 3
```

```python
# figure size
plt.figure(figsize=(8,4))

# plot function
plt.plot(x, y1, 'g--', label='Squared') 
plt.plot(x, y2, 'rs', label='Cubic')


# title
plt.title("Karlos Garcia")

# axis
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
# show
plt.show()
```

6. Use the data below and plot the in a barchart 

```python
# data
x = np.arange(8,19)
y = x ** 2
```

```python
# figure size
plt.figure(figsize=(8,4))

# plot function
plt.bar(x,y)

# title
plt.title("Karlos Garcia")

# axis
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
# show
plt.show()
```

7. Create the **x** and **y** variable with 1000 data points derived from the normal distribution rispectively.
 * Plot them in a scatter plot

```python
# data
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
# set fig size
plt.figure(figsize=(8,4))

# plot
plt.scatter(x,y, c='r')

# show
plt.show()
```

8. Use the following data and plot two subplots. One with squared relation and one with linear relation.

```python
# data
x = np.arange(1,1000)
y = x
```

```python
# set fig and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

# Plot 1
axes[0].plot(x, y**2, 'r--')


# Plot 2
axes[1].plot(x, y, 'g')

# show
plt.show()
```
