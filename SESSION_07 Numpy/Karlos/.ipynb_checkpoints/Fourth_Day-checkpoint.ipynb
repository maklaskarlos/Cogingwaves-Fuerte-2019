```python
# import numpy
import numpy as np

# here we import a specific package
import matplotlib.pyplot as plt

# this code is needed to visualize the pots
%matplotlib inline
```

```python
x = np.arange(2, 21)
y = 2 * x + 5
```

```python
plt.figure(figsize=(9,4))

plt.plot(x, y, "g-_", label="My Data")

plt.title("First Plot")

plt.xlabel("Value of x")
plt.ylabel("Value of y")

plt.legend()
plt.show()
```

```python
x= np.arange(1,11)
y= np.exp(x)
```

```python
plt.figure(figsize=(8,5))

plt.plot(x, y, "g", label= "My second Data")

plt.title("Second Plot")
plt.xlabel("This is the x label")
plt.ylabel("This is the x label")
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = x ** 2
```

```python
plt.figure(figsize=(8,4))

plt.bar(x,y)

plt.title("This is a barplot")
plt.xlabel("Only x label")
plt.ylabel("This is y")

plt.show()
```

```python
var = np.random.randn(100000)

plt.figure(figsize=(8,5))

plt.hist(var, bins=100)

plt.show()
```

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
plt.figure(figsize=(10,5))

plt.scatter(x,y, c="r")
plt.grid()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10,5))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

axes[0].plot(x, y**2, "ro")
axes[0].set_title("Karlos")

axes[1].plot(x,y, "o")
axes[1].set_title("Garcia")

plt.show()
```
