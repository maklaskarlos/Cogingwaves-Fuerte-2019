```python
# import numpy
import numpy as np

# here we import a specific package
import matplotlib.pyplot as plt

# this code is needed to visualize the pots
%matplotlib inline
```

```python
x = np.arange(2, 21)
y = 2 * x + 5
```

```python
plt.figure(figsize=(9,4))

plt.plot(x, y, "g-_", label="My Data")

plt.title("First Plot")

plt.xlabel("Value of x")
plt.ylabel("Value of y")

plt.legend()
plt.show()
```

```python
x= np.arange(1,11)
y= np.exp(x)
```

```python
plt.figure(figsize=(8,5))

plt.plot(x, y, "g", label= "My second Data")

plt.title("Second Plot")
plt.xlabel("This is the x label")
plt.ylabel("This is the x label")
plt.legend()
plt.show()
```
